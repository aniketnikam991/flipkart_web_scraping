```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://www.flipkart.com/search?q=smartphone%20under%2020000&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"
```

```python
url
```

```python
response = requests.get(url)
```

```python
response
```

```python
soup = BeautifulSoup(response.text, "lxml")
soup
```

```python
box = soup.find('div',class_="QSCKDh dLgFEE")
```

```python
names = box.find_all("div",class_ = "RG5Slk")
```

```python
reviews = box.find_all('div',class_="MKiFS6")
```

```python
names
```

```python
reviews
```

```python
product_name = []
product_review = []
```

```python
for i in names:
    name = i.text
    product_name.append(name)
```

```python
for i in reviews:
    review = i.text
    product_review.append(review)
```

```python
product_name
```

```python
product_review
```

```python
df = pd.DataFrame({"Product Name":product_name,"Review":product_review})
df
```

```python
df.to_csv('Flipkart_Budget_Phones.csv')
```
